```python
import pandas as pd
import seaborn as sns
```

```python
## simple linear regression
dict_regression = {'time_in_corp': [1, 3, 6, 9, 10, 14, 18], 'salary': [1500, 3000, 4500, 6000, 7000, 8500, 10000]}
```

```python
df_simple_regression = pd.DataFrame.from_dict(dict_regression)
```

```python
sns.regplot(data=df_simple_regression, x='time_in_corp', y='salary')
```
